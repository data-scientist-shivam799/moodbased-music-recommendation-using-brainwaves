```python
import pandas as pd
```

```python
df=pd.read_csv('D:/Study/Brainwaves/emotiv_eeg_1.csv')
df.head()
```

```python
# df['AF3_Attention/Meditation']=df['AF3_Attention/Meditation'].astype('float64')
df=df.iloc[:, 0:14].values
df
```

# Feature Scaling

```python
# performing preprocessing part
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

df=sc.fit_transform(df)
df=pd.DataFrame(df,columns=['AF3_Attention/Meditation','AF4_Attention/Meditation','F3_Eye Blinking_and_Facial_Expressions',
                         'F4_Eye Blinking_and_Facial_Expressions','F7_Eye Blinking_and_Facial_Expressions',
                         'F8_Eye Blinking_and_Facial_Expressions','FC5_Motor_Imagery_and_Spatial_Navigation',
                         'FC6_Motor_Imagery_and_Spatial_Navigation','T7_Auditory_Processing_and_Emotion',
                         'T8_Auditory_Processing_and_Emotion','O1_Visual_Processing','O2_Visual_Processing',
                         'P7_Visual_Processing','P8_Visual_Processing'])
df.describe()
```


# Categorizing Features

```python
df_Attention_Meditation_Comb=df[['AF3_Attention/Meditation','AF4_Attention/Meditation']]
df_Attention_Meditation_Comb.head(2)
```

```python
df_Eye_Blinking_and_Facial_Expressions_Comb=df[['F3_Eye Blinking_and_Facial_Expressions','F4_Eye Blinking_and_Facial_Expressions',
                                          'F7_Eye Blinking_and_Facial_Expressions','F8_Eye Blinking_and_Facial_Expressions']]
df_Eye_Blinking_and_Facial_Expressions_Comb.head(2)
```

```python
df_Motor_Imagery_and_Spatial_Navigation_Comb=df[['FC5_Motor_Imagery_and_Spatial_Navigation',
                                                  'FC6_Motor_Imagery_and_Spatial_Navigation']]
df_Motor_Imagery_and_Spatial_Navigation_Comb.head(2)
```

```python
df_Auditory_Processing_and_Emotion_Comb=df[['T7_Auditory_Processing_and_Emotion','T8_Auditory_Processing_and_Emotion']]
df_Auditory_Processing_and_Emotion_Comb.head(2)
```

```python
df_Visual_Processing_Comb=df[['O1_Visual_Processing','O2_Visual_Processing','P7_Visual_Processing','P8_Visual_Processing']]
df_Visual_Processing_Comb.head(2)
```

```python
df_Visual_Processing_Comb['P8_Visual_Processing'].isnull().sum()
```

# Applying PCA To Different Categories

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components = 1)
```

```python
df_Attention_Meditation=pca.fit_transform(df_Attention_Meditation_Comb)
explained_variance = pca.explained_variance_ratio_
print('explained_variance is',explained_variance)
df_Attention_Meditation=pd.DataFrame(df_Attention_Meditation,columns=['Attention_Meditation'])
df_Attention_Meditation.head(2)
```

```python
df_Eye_Blinking_and_Facial_Expressions=pca.fit_transform(df_Eye_Blinking_and_Facial_Expressions_Comb)
explained_variance = pca.explained_variance_ratio_
print('explained_variance is',explained_variance)
df_Eye_Blinking_and_Facial_Expressions=pd.DataFrame(df_Eye_Blinking_and_Facial_Expressions,
                                                    columns=['Eye_Blinking_and_Facial_Expressions'])
df_Eye_Blinking_and_Facial_Expressions.head(2)
```

```python
df_Motor_Imagery_and_Spatial_Navigation=pca.fit_transform(df_Motor_Imagery_and_Spatial_Navigation_Comb)
explained_variance = pca.explained_variance_ratio_
print('explained_variance is',explained_variance)
df_Motor_Imagery_and_Spatial_Navigation=pd.DataFrame(df_Motor_Imagery_and_Spatial_Navigation,
                                                    columns=['Motor_Imagery_and_Spatial_Navigation'])
df_Motor_Imagery_and_Spatial_Navigation.head(2)
```

```python
df_Auditory_Processing_and_Emotion=pca.fit_transform(df_Auditory_Processing_and_Emotion_Comb)
explained_variance = pca.explained_variance_ratio_
print('explained_variance is',explained_variance)
df_Auditory_Processing_and_Emotion=pd.DataFrame(df_Auditory_Processing_and_Emotion,
                                                    columns=['Auditory_Processing_and_Emotion'])
df_Auditory_Processing_and_Emotion.head(2)
```

```python
# dealing with NaN values
import numpy as np
np.any(np.isnan(df_Visual_Processing_Comb))

df_Visual_Processing=pca.fit_transform(df_Visual_Processing_Comb)
explained_variance = pca.explained_variance_ratio_
print('explained_variance is',explained_variance)
df_Visual_Processing=pd.DataFrame(df_Visual_Processing,columns=['Visual_Processing'])
df_Visual_Processing.head(2)
```

# Merging the five PCA components

```python
COUNTER=pd.DataFrame(df['COUNTER'].values,columns=['COUNTER'])
new_df=pd.concat([df_Attention_Meditation,df_Eye_Blinking_and_Facial_Expressions,df_Motor_Imagery_and_Spatial_Navigation,
                 df_Auditory_Processing_and_Emotion,df_Visual_Processing,COUNTER],axis=1)
new_df.head()
```
