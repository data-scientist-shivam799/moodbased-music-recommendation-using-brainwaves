```python
def interval(dataset,num):
    final=pd.DataFrame(columns=['Attention_Meditation','Eye_Blinking_and_Facial_Expressions',
                               'Motor_Imagery_and_Spatial_Navigation','Auditory_Processing_and_Emotion','Visual_Processing'])
    start=0
    freq=num
    batch_size=len(dataset)//num 
    print('Batch Size:',batch_size)
    for i in range(batch_size):
        final=final.append(dataset[start:freq].mode(),ignore_index = True)
#         print("Processing batch of "+str(start)+" to "+str(freq))
        if i>batch_size:
            break
        else:
            start+=num
            freq+=num
    return final
```

```python
df_eeg['Activty_Score']=df_eeg.sum(axis=1)
df_eeg.head()
```

```python
# Finding activity 
def activity_score(num):
    if num<=-3.0:
        return 'very low'
    elif num>-3.0 and num<=0:
        return 'low'
    elif num==0.0:
        return 'neutral'
    elif num>0 and num<=3.0:
        return 'high'
    elif num>3.0:
        return 'very high'
    else:
        return 'Enter a valid value'
    
activity_score(df_eeg['Activty_Score'][2])
```

Activity_Score below -3.0 -> 'very low' <br>
Activity_Score above -3.0 but below 0.0 -> 'low' <br>
Activity_Score 0.0 -> 'neutral'<br>
Activity_Score above 0.0 but below 3.0 -> 'high'<br>
Activity_Score above 3.0 -> 'very high'

```python
import numpy as np

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import euclidean_distances
from scipy.spatial.distance import cdist

from collections import defaultdict

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
class Music_Recommendation:
    def __init__(self,eeg_data,music_data,activity=activity_score(df_eeg['Activty_Score'][2])):
        self.eeg_data=eeg_data
        self.music_data=music_data
        self.activity=activity
        print("Recommendation engine started...")
    def categorized_music_dataset(self,activity):
        # Case 1: Very low activity
        if self.activity=='very low':
            self.music_data=self.music_data[self.music_data['tempo']<=93]
            self.music_data=self.music_data[self.music_data['danceability']<=0.415]
            self.music_data=self.music_data[self.music_data['energy']<=0.255]
        # Case 2: low activity
        elif self.activity=='low':
            self.music_data=self.music_data[self.music_data['tempo']>93]
            self.music_data=self.music_data[self.music_data['tempo']<=115]
            self.music_data=self.music_data[self.music_data['danceability']>0.415]
            self.music_data=self.music_data[self.music_data['danceability']<=0.548]
            self.music_data=self.music_data[self.music_data['energy']>0.255]
            self.music_data=self.music_data[self.music_data['energy']<=0.471]
        # Case 3: high activity
        elif self.activity=='high':
            self.music_data=self.music_data[self.music_data['tempo']>115]
            self.music_data=self.music_data[self.music_data['tempo']<=131]
            self.music_data=self.music_data[self.music_data['danceability']>0.548]
            self.music_data=self.music_data[self.music_data['danceability']<=0.642]
            self.music_data=self.music_data[self.music_data['energy']>0.471]
            self.music_data=self.music_data[self.music_data['energy']<=0.656]
        # Case 4: Very high activity
        elif self.activity=='very high':
            self.music_data=self.music_data[self.music_data['tempo']>131]
            self.music_data=self.music_data[self.music_data['danceability']>0.642]
            self.music_data=self.music_data[self.music_data['energy']>0.656]
        else:
            print('Pass appropriate parameter -> very low, low, high and very high')
        return self.music_data
    
    def recommend_music(self):
        data=self.categorized_music_dataset(self.activity)
#         print(data.head(1))

        song_cluster_pipeline = Pipeline([('scaler', StandardScaler()), 
                                          ('kmeans', KMeans(n_clusters=20, 
                                           verbose=False))
                                         ], verbose=False)

        X = data.select_dtypes(np.number)
        number_cols = list(X.columns)
        song_cluster_pipeline.fit(X)
        song_cluster_labels = song_cluster_pipeline.predict(X)
        data['cluster_label'] = song_cluster_labels



        sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="f0332ee7ce1d4d95a52d7cd899b9b0f0",
                                                                   client_secret="34f6805517af450b8cc78d633ff83ff3"))

        def find_song(name, year):
            song_data = defaultdict()
            results = sp.search(q= 'track: {} year: {}'.format(name,year), limit=1)
            if results['tracks']['items'] == []:
                return None

            results = results['tracks']['items'][0]
            track_id = results['id']
            audio_features = sp.audio_features(track_id)[0]

            song_data['name'] = [name]
            song_data['year'] = [year]
            song_data['explicit'] = [int(results['explicit'])]
            song_data['duration_ms'] = [results['duration_ms']]
            song_data['popularity'] = [results['popularity']]

            for key, value in audio_features.items():
                song_data[key] = value

            return pd.DataFrame(song_data)
        
        
        
        number_cols = ['valence', 'year', 'acousticness', 'danceability', 'energy',
         'instrumentalness', 'key', 'liveness', 'loudness', 'popularity', 'speechiness', 'tempo']


        def get_song_data(song, spotify_data):

            try:
                song_data = spotify_data[(spotify_data['name'] == song['name']) 
                                        & (spotify_data['year'] == song['year'])].iloc[0]
                return song_data

            except IndexError:
                return find_song(song['name'], song['year'])


        def get_mean_vector(song_list, spotify_data):

            song_vectors = []

            for song in song_list:
                song_data = get_song_data(song, spotify_data)
                if song_data is None:
                    print('Warning: {} does not exist in Spotify or in database'.format(song['name']))
                    continue
                song_vector = song_data[number_cols].values
                song_vectors.append(song_vector)  

            song_matrix = np.array(list(song_vectors))
            return np.mean(song_matrix, axis=0)


        def flatten_dict_list(dict_list):

            flattened_dict = defaultdict()
            for key in dict_list[0].keys():
                flattened_dict[key] = []

            for dictionary in dict_list:
                for key, value in dictionary.items():
                    flattened_dict[key].append(value)

            return flattened_dict


        def recommend_songs( song_list, spotify_data, n_songs=10):

            metadata_cols = ['name', 'year', 'artists']
            song_dict = flatten_dict_list(song_list)

            song_center = get_mean_vector(song_list, spotify_data)
            scaler = song_cluster_pipeline.steps[0][1]
            scaled_data = scaler.transform(spotify_data[number_cols])
            scaled_song_center = scaler.transform(song_center.reshape(1, -1))
        #     distances = cdist(scaled_song_center, scaled_data, 'cosine')
            distances=euclidean_distances(scaled_song_center,scaled_data)
            index = list(np.argsort(distances)[:, :n_songs][0])

            rec_songs = spotify_data.iloc[index]
            rec_songs = rec_songs[~rec_songs['name'].isin(song_dict['name'])]
            return rec_songs[metadata_cols]
        
#         return recommend_songs([{'name': 'A Thousand Years', 'year':2011}],  data)
#         return recommend_songs([{'name': 'Overture', 'year':1921}],  data)   # high
        return recommend_songs([{'name': 'Rákóczy March', 'year':1921}],  data)   # very high
```
